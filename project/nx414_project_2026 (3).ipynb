## 0.2 Why task-optimized models?

Task-optimized neural networks are among the most useful current **in-silico models of sensory cortex**. The central idea is simple: instead of hand-designing a model to mimic every biological detail, we optimize a model to perform a meaningful visual task and then ask whether its internal representations resemble those found in the brain. This approach has been highly influential because models trained to solve vision tasks often develop representations that predict activity along the visual hierarchy surprisingly well.

These models are useful scientifically because they provide **testable computational hypotheses**. If a model layer predicts neural responses well, that does not mean the brain literally implements the same mechanism, but it does suggest that the layer may encode information in a similar format or at a similar level of abstraction. Brain–model alignment is therefore a way to ask not just whether a model is accurate on a task, but whether it organizes visual information in a brain-relevant way.

## 0.3 Why compare multiple modalities?

A single recording modality gives only a partial view of neural computation. In this project, you will work with **electrophysiology, EEG, and fMRI**, which differ in temporal resolution, spatial resolution, and what exactly is measured. Looking across modalities helps you see which conclusions are robust and which depend on the measurement scale.

### Noise ceilings

Noise ceilings are stored per target:

- per neuron for **TVSD**,
- per channel × time point for **EEG2**,
- per voxel for **NSD**.

They are stored as **percent reliability**.  
To convert them to the range `[0, 1]`, divide by `100`.

In this project, the provided noise ceilings are mainly intended for **predictive metrics** such as **Pearson correlation** and **explained variance**. They reflect the reliability of the neural responses and therefore define an upper bound on how well any model can predict those responses.

When you compute predictive metrics, you should apply a noise ceiling correction to account for this upper bound. The standard idea is simple: divide the raw predictive score by the corresponding noise ceiling value for that target.

The provided noise ceilings are defined for **explained variance**. If you want to apply the same logic to **Pearson correlation**, first convert the explained-variance ceiling into a correlation ceiling by taking the element-wise square root, and then divide the raw correlation by that quantity. For a more detailed discussion of noise ceiling correction, see van Bree et al. (2025).

You may therefore report both **raw** and **noise-ceiling-corrected** predictive scores where appropriate.


For example, if the provided ceiling is an explained-variance reliability estimate, you can compute a noise-corrected Pearson correlation as:

```python
r_nc = r / np.sqrt(ev_ceiling)
```

where `r` is the raw Pearson correlation and `ev_ceiling` is the explained-variance ceiling expressed on the range `[0, 1]`.

By contrast, **RSA** and **CKA** should typically be reported as **raw values** in this project. Noise ceilings for representational similarity metrics require a different methodology and are **not** provided here.


Do **not** apply this correction directly to **RSA** or **CKA**.

**Define utility functions**

````python
def explore_h5(path, indent=0):
    """Explore h5 file in detail."""
    with h5py.File(path, 'r') as f:
        def print_structure(name, obj):
            print(' ' * indent + name, ':', type(obj).__name__, end='')
            if isinstance(obj, h5py.Dataset):
                print(f'  shape={obj.shape}, dtype={obj.dtype}')
            else:
                print()
        f.visititems(print_structure)

def load_h5_attrs(path):
    """
    Recursively collect all HDF5 attributes from every group and dataset.
    Returns a nested dict: {group_or_dataset_path: {attr_name: attr_value, ...}, ...}
    Root-level attributes are stored under the key '/'.
    """
    attrs = {}
    with h5py.File(path, 'r') as f:
        if f.attrs:
            attrs['/'] = dict(f.attrs)
        def _collect(name, obj):
            if obj.attrs:
                attrs[name] = dict(obj.attrs)
        f.visititems(_collect)
    return attrs

def load_neural_IDs(path, split='train', subjects=None):
    """    Load stimulus IDs for a given split (train or test).
    Handles both single (TVSD, EEG2) and per-subject (NSD) structures.
    """
    with h5py.File(path, 'r') as f:
        ids_key = f'{split}/stimulus_ids'
        obj = f[ids_key]
        
        if isinstance(obj, h5py.Dataset):
            # TVSD / EEG2: single stimulus_ids array
            return obj[:]   

        elif isinstance(obj, h5py.Group):
            # NSD: one array per subject
            if subjects is None:
                subjects = list(obj.keys())
            return {sub: obj[sub][:] for sub in subjects}

def load_feature_IDs(path):
    """Load stimulus IDs from a feature h5 file."""
    with h5py.File(path, 'r') as f:
        return f['ids'][:]

def match_feature_idx(feature_file, neural_ids):
    """
    Build feature indices aligned to neural_ids,
    following efficient 5 read procedure.
    """
    # Step 1: build index array

    # load all stimulus IDs from feature file
    with h5py.File(feature_file, 'r') as f:
        feat_ids = f['ids'][:]
        # Builds a dictionary: {id: index} for fast lookup
        id_to_feat_idx = {id_: i for i, id_ in enumerate(feat_ids)} 
        # Ror each neural stimulus ID, find its position in the feature stimulu ID array
        feat_idx       = np.array([id_to_feat_idx[x] for x in neural_ids]) 
        
        # Step 2: sort indices for efficient HDF5 reads
        sort_order     = np.argsort(feat_idx)
        feat_idx_sorted = feat_idx[sort_order]
    
        # Step 3: load only required rows (sorted)
        features_sorted = f['ids'][feat_idx_sorted]
        
    # Step 4: restore original order --> = matching the neural ids
    restore_order = np.empty_like(sort_order)
    restore_order[sort_order] = np.arange(len(sort_order))
    matched_ids = features_sorted[restore_order]
    
    return matched_ids
````

````python
import numpy as np


def compute_ceiling_variancebased(responses: np.ndarray, nan_policy: str = 'omit') -> np.ndarray:
    """
    Noise ceiling per unit using the method described in the NSD paper
    (Allen et al., 2021 / 2022 style variance-based estimator).

    Steps:
      1) z-score across stimuli (axis=1) for each (unit, rep) -> total var ≈ 1
      2) estimate noise variance across repetitions (axis=2), then average across stimuli
      3) signal variance = 1 - noise_var
      4) reliability (percent) for finite repeats:
             nc = 100 * (snr / (snr + 1 / n_reps))

    Parameters
    ----------
    responses : np.ndarray
        Shape (n_units, n_stimuli, n_reps) or
        (n_channels, n_timepoints, n_stimuli, n_reps).
    nan_policy : {'propagate', 'raise', 'omit'}, default='omit'
        Passed to the z-scoring logic when handling NaNs.

    Returns
    -------
    np.ndarray
        Per-unit noise ceilings in percent with shape (n_units,) or
        (n_channels, n_timepoints), depending on your implementation.
    """

    if responses.ndim == 4:
        n_channels, n_timepoints, n_stimuli, n_reps = responses.shape
        # process each (channel, timepoint) unit independently
        X = responses.reshape(n_channels * n_timepoints, n_stimuli, n_reps)

    # TVSD: (n_units, n_stimuli, n_reps)
    elif responses.ndim == 3:
        n_units, n_stimuli, n_reps = responses.shape
        X = responses.copy()

    else:
        raise ValueError(f"Expected 3D or 4D array, got shape {responses.shape}")

    n_units, n_stimuli, n_reps = X.shape

    # Step 1: z-score across stimuli for each (unit, rep)
    if nan_policy == 'omit':
        mean = np.nanmean(X, axis=1, keepdims=True)
        std  = np.nanstd( X, axis=1, keepdims=True)
    elif nan_policy == 'propagate':
        mean = X.mean(axis=1, keepdims=True)
        std  = X.std( axis=1, keepdims=True)
    elif nan_policy == 'raise':
        if np.any(np.isnan(X)):
            raise ValueError("NaNs found in responses and nan_policy='raise'")
        mean = X.mean(axis=1, keepdims=True)
        std  = X.std( axis=1, keepdims=True)
    else:
        raise ValueError(f"Unknown nan_policy: {nan_policy!r}")

    std = np.where(std == 0, np.nan, std)
    Z = (X - mean) / std                              # (n_units, n_stimuli, n_reps)

    # Step 2: noise variance across reps, averaged across stimuli
    if nan_policy == 'omit':
        noise_var = np.nanmean(np.nanvar(Z, axis=2, ddof=1), axis=1)
    else:
        noise_var = np.mean(np.var(Z, axis=2, ddof=1), axis=1)

    # Step 3: signal variance
    signal_var = np.maximum(1.0 - noise_var, 0)

    # Step 4: finite-repeat correction -> (n_units,)
    nc = 100.0 * (signal_var / (signal_var + 1.0 / n_reps))

    # Reshape output to match input structure
    if responses.ndim == 4:
        return nc.reshape(n_channels, n_timepoints)   # EEG: (n_channels, n_timepoints)
    else:
        return nc                                      # TVSD: (n_units,)


def compute_ceiling_splithalf(
    responses: np.ndarray,
    folds: int = 10,
    seed: int = 0,
    spearman_brown: bool = True,
    equalize_halves: bool = True,
    clip_folds: bool = False
) -> np.ndarray:
    """
    Split-half reliability per unit (voxel / channel / channel*timepoint).
    You can refer to van Bree et al. (2025) for mathematical details.
    
    Steps:
      1) For each fold, randomly split repetitions into two halves.
      2) Average responses within each half and compute Pearson correlation across stimuli.
      3) Optionally apply Spearman-Brown correction to each fold's correlation.
      4) Average across folds to get a final reliability estimate.

    Parameters
    ----------
    responses : np.ndarray
        Shape (n_units, n_stimuli, n_reps) or
        (n_channels, n_timepoints, n_stimuli, n_reps).
        The last axis corresponds to repetitions / trials.
    folds : int, default=10
        Number of random split-halves to sample.
    seed : int, default=0
        Base RNG seed; each fold may use seed + fold_idx.
    spearman_brown : bool, default=True
        Apply Spearman-Brown correction:
            r_sb = 2r / (1 + r)
    equalize_halves : bool, default=True
        If True, use equal-sized halves and drop one trial if n_reps is odd.
        If False, the second half may be larger by one trial.
    clip_folds : bool, default=False
        If True, clip reliability values after correction.

    Returns
    -------
    np.ndarray
        Array of shape (n_units) or (n_channels, n_timepoints).
    """
    # EEG: (n_channels, n_timepoints, n_stimuli, n_reps)
    if responses.ndim == 4:
        n_channels, n_timepoints, n_stimuli, n_reps = responses.shape
        X = responses.reshape(n_channels * n_timepoints, n_stimuli, n_reps)

    # TVSD: (n_units, n_stimuli, n_reps)
    elif responses.ndim == 3:
        n_units, n_stimuli, n_reps = responses.shape
        X = responses.copy()

    else:
        raise ValueError(f"Expected 3D or 4D array, got shape {responses.shape}")

    n_units, n_stimuli, n_reps = X.shape
    half = n_reps // 2  # size of each half (drops one rep if n_reps is odd + equalize_halves)

    fold_corrs = np.zeros((folds, n_units))  # (folds, n_units)

    for fold in range(folds):
        rng = np.random.default_rng(seed + fold)
        perm = rng.permutation(n_reps)

        if equalize_halves:
            idx_a = perm[:half]
            idx_b = perm[half:2*half]   # drop last rep if n_reps is odd
        else:
            idx_a = perm[:half]
            idx_b = perm[half:]         # second half may be larger by one

        # Average across reps within each half -> (n_units, n_stimuli)
        half_a = X[:, :, idx_a].mean(axis=2)
        half_b = X[:, :, idx_b].mean(axis=2)

        # Step 2: Pearson correlation across stimuli for each unit
        # z-score each half across stimuli, then dot product
        def zscore(arr):
            mu  = arr.mean(axis=1, keepdims=True)
            std = arr.std( axis=1, keepdims=True)
            std = np.where(std == 0, np.nan, std)
            return (arr - mu) / std

        za = zscore(half_a)  # (n_units, n_stimuli)
        zb = zscore(half_b)  # (n_units, n_stimuli)

        r = np.nanmean(za * zb, axis=1)  # (n_units,) — Pearson r per unit

        # Step 3: Spearman-Brown correction
        if spearman_brown:
            r = 2 * r / (1 + r)

        if clip_folds:
            r = np.clip(r, 0, 1)

        fold_corrs[fold] = r

    # Step 4: average across folds -> (n_units,)
    nc = np.nanmean(fold_corrs, axis=0) * 100

    if responses.ndim == 4:
        return nc.reshape(n_channels, n_timepoints)
    else:
        return nc
````

## 2.1 Representational alignment: RSA

RSA stands for representational similarity analysis. It is one of the most widely used analyses in fMRI and model–brain alignment research. It compares the geometry of two representational spaces through their representational dissimilarity matrices (RDMs). Given two response matrices, `X` and `Y`, with rows corresponding to the same stimuli, we first compute an RDM for each matrix using correlation distance:

$$
D^X_{ij} = 1 - \mathrm{corr}(X[i,:], X[j,:]),
\qquad
D^Y_{ij} = 1 - \mathrm{corr}(Y[i,:], Y[j,:]),
$$

for stimulus pairs $i \neq j$.

We then vectorize the upper triangle of each RDM and compute RSA as the Spearman correlation between these two vectors:

$$
\mathrm{RSA}(X, Y)
=
\rho_{\mathrm{Spearman}}
\left(
\mathrm{vec}(D^X),\,
\mathrm{vec}(D^Y)
\right).
$$

In this project, `X` will usually denote model features from one candidate layer, and `Y` will denote neural responses from one dataset, ROI, subject, or time slice, depending on the analysis.

- Implement RSA between two representation matrices.
- Support at least one dissimilarity measure and one similarity measure.
- Use your implementation to compare model layers to neural responses.

````python
from typing import Literal
import numpy as np

class RepresentationalSimilarityAnalysis:
    """
    Representational Similarity Analysis (RSA).

    Given two representation matrices X and Y with the same number of conditions
    (rows), RSA:

    1. Computes a Representational Dissimilarity Matrix (RDM) for each:
       RDM_X[i, j] = dissimilarity(x_i, x_j)
       RDM_Y[i, j] = dissimilarity(y_i, y_j)

    2. Flattens the upper triangles of both RDMs and computes a correlation
       between them (Pearson or Spearman).
    """

    def __init__(
        self,
        dissimilarity: Literal["correlation", "euclidean", "cosine"] = "correlation",
        similarity_metric: Literal["pearson", "spearman"] = "spearman",
    ):
        ### TODO
        self.dissimilarity = dissimilarity
        self.similarity_metric = similarity_metric


    def __call__(self, X: np.ndarray, Y: np.ndarray) -> float:
        """
        Compute RSA similarity between X and Y.

        Parameters
        ----------
        X, Y : np.ndarray
            Arrays of shape (n_conditions, ...) that may need to be flattened
            along feature dimensions.

        Returns
        -------
        rsa_similarity : float
            Correlation between the vectorized upper triangles of the two RDMs.
        """
        return self.forward(X, Y)

    def forward(self, X: np.ndarray, Y: np.ndarray) -> float:
        ### TODO
        #Flattening extra features to handle 2D matrices
        X_flat = X.reshape(X.shape[0], -1).astype(np.float64)
        Y_flat = Y.reshape(Y.shape[0], -1).astype(np.float64)

        rdm_x = self.compute_rdm(X_flat)
        rdm_y = self.compute_rdm(Y_flat)
        return self.compare_rdms(rdm_x, rdm_y)

    def compute_rdm(self, X: np.ndarray) -> np.ndarray:
        """
        Compute the Representational Dissimilarity Matrix (RDM)
        for a given representation matrix X.

        Parameters
        ----------
        X : np.ndarray
            Array of shape (n_conditions, n_features).

        Returns
        -------
        rdm : np.ndarray
            Array of shape (n_conditions, n_conditions) with pairwise dissimilarities.
        """
        ### TODO
        X_flat = X.reshape(X.shape[0], -1).astype(np.float64)

        #cdist computes n*(n-1)/2 pairwise distances efficiently, "correlation" distance: D_ij = 1 - pearson(x_i, x_j),
        return cdist(X_flat, X_flat, metric=self.dissimilarity)
        

    def compare_rdms(self, rdm1: np.ndarray, rdm2: np.ndarray) -> float:
        """
        Compare two RDMs by correlating their upper triangles.
        """
        ### TODO
        n   = rdm1.shape[0]
        idx = np.triu_indices(n, k=1) #selects elements strictly above the diagonal, no self distances 
        vec1 = rdm1[idx]
        vec2 = rdm2[idx]

        #computes correlation between the 2 upper triangles of the RDMs
        if self.similarity_metric == "spearman":
            r, _ = spearmanr(vec1, vec2)
        else:
            r, _ = pearsonr(vec1, vec2)

        return float(r)
````

## 2.2 Representational alignment: unbiased linear CKA

CKA stands for centered kernel alignment. It is commonly used in interpretability and representation analysis to test how strongly the internal computations of two systems align. As a second mapping-free alignment metric, we want to compute unbiased linear centered kernel alignment (CKA) between model features and neural responses. Let

$$
X \in \mathbb{R}^{n \times d}, \qquad
Y \in \mathbb{R}^{n \times p},
$$

where both matrices are measured on the same $n$ stimuli. We form linear Gram matrices

$$
K = XX^\top,
\qquad
L = YY^\top.
$$

We then estimate dependence using the unbiased (U-statistic) HSIC estimator, $\mathrm{HSIC}_u(K, L)$, and define CKA as

$$
\mathrm{CKA}(X, Y)
=
\frac{\mathrm{HSIC}_u(K, L)}
{\sqrt{\mathrm{HSIC}_u(K, K)\,\mathrm{HSIC}_u(L, L)}}.
$$

Like RSA, CKA compares representational structure directly without fitting a predictive mapping. In this notebook, `X` and `Y` again refer to aligned model and neural response matrices evaluated on the same set of stimuli.

- Implement **unbiased linear CKA** only.
- Use your implementation to compare model layers to neural responses.

### Starter code

````python
import numpy as np

class CenteredKernelAlignment:
    """
    Unbiased linear CKA only.

    Parameters
    ----------
    eps : float
        Small constant for numerical stability.
    dtype : np.dtype
        Data type used for computations.
    """

    def __init__(
        self,
        eps: float = 1e-8,
        dtype: np.dtype = np.float64,
    ):
        self.eps = eps
        self.dtype = dtype

    def __call__(self, X: np.ndarray, Y: np.ndarray) -> float:
        return self.forward(X, Y)

    def forward(self, X: np.ndarray, Y: np.ndarray) -> float:
        X = np.asarray(X).astype(self.dtype)
        Y = np.asarray(Y).astype(self.dtype)

        if X.shape[0] != Y.shape[0]:
            raise ValueError(
                f"Batch sizes must match along axis 0: {X.shape[0]} vs {Y.shape[0]}"
            )

        # Flatten to (n_samples, n_features)
        X = X.reshape(X.shape[0], -1)
        Y = Y.reshape(Y.shape[0], -1)

        return self._unbiased_linear_cka(X, Y)

    def _unbiased_linear_hsic(self, X: np.ndarray, Y: np.ndarray) -> float:
        """
        Unbiased HSIC estimator for the linear kernel (Song et al. 2012, Eq. 5).

        X : [n, d_x]
        Y : [n, d_y]
        """
        n = X.shape[0]
        if n < 4:
            raise ValueError(f"Need at least 4 samples for unbiased HSIC; got {n}")

        # Linear Gram matrices
        K = X @ X.T
        L = Y @ Y.T

        # K_tilde and L_tilde: diagonal entries set to zero
        np.fill_diagonal(K, 0)
        np.fill_diagonal(L, 0)

        trace_KL = (K * L).sum()
        correction = K.sum() * L.sum() / ((n - 1) * (n - 2))
        cross_term = (2.0 / (n - 2)) * (K.sum(axis=1) @ L.sum(axis=1))

        return (trace_KL + correction - cross_term) / (n * (n - 3))



    def _unbiased_linear_cka(self, X: np.ndarray, Y: np.ndarray) -> float:
        """
        Unbiased linear CKA:

            CKA_unb(X, Y) =
                HSIC_unb(X, Y) / sqrt(HSIC_unb(X, X) * HSIC_unb(Y, Y))
        """
        hsic_xy = self._unbiased_linear_hsic(X, Y)
        hsic_xx = self._unbiased_linear_hsic(X, X)
        hsic_yy = self._unbiased_linear_hsic(Y, Y)

        denom = np.sqrt(max(hsic_xx * hsic_yy, 0.0)) + self.eps # max(..., 0.0) against tiny negative values due to floating point errors
        return float(hsic_xy / denom)
````

````python
def get_feature_layers(feature_path):
    """Return all feature layers stored in the feature h5 file."""
    layers = []
    with h5py.File(feature_path, "r") as f:
        for k in f.keys(): 
            if k == "ids":
                continue
            if isinstance(f[k], h5py.Group):
                # layers nested inside a group (e.g. "features/layer1-0")
                for sub_k in f[k].keys():
                    if isinstance(f[k][sub_k], h5py.Dataset):
                        layers.append(f"{k}/{sub_k}")
            else:
                # flat structure (e.g. "layer1-0")
                layers.append(k)
    return layers

def load_features_aligned(feature_path, layer, neural_ids):
    with h5py.File(feature_path, "r") as f:
        feat_ids = f["ids"][:].astype(str)          
        neural_ids_str = np.array(neural_ids).astype(str)

        feat_id_to_idx = {id_: i for i, id_ in enumerate(feat_ids)}

       
        missing = [id_ for id_ in neural_ids_str if id_ not in feat_id_to_idx]
        if missing:
            raise KeyError(f"{len(missing)} neural IDs not find: {missing[:5]}...")

        feat_idx = np.array([feat_id_to_idx[id_] for id_ in neural_ids_str])

        sort_order = np.argsort(feat_idx)
        X_sorted = f[layer][feat_idx[sort_order]]

    restore_order = np.empty_like(sort_order)
    restore_order[sort_order] = np.arange(len(sort_order))
    return X_sorted[restore_order]

def compute_layerwise_scores_avg(feature_path, subjects_data, dataset, target, model):
    """Compute RSA/CKA per subject/monkey separately, then average the scores across subjects."""
    layers = get_feature_layers(feature_path)
    all_subject_dfs = []

    for sub_idx, (ids, Y) in enumerate(subjects_data):
        rows = []
        for layer in tqdm(layers, desc=f"{dataset} | {target} | {model} | sub {sub_idx+1}/{len(subjects_data)}", leave=False):
            X = load_features_aligned(feature_path, layer, ids)
            rows.append({
                "layer": layer,
                "RSA": rsa(X, Y),
                "CKA": cka(X, Y)
            })
        all_subject_dfs.append(pd.DataFrame(rows))

    avg = (
        pd.concat(all_subject_dfs)
        .groupby("layer")[["RSA", "CKA"]]
        .mean()
        .reset_index()
    )
    avg["dataset"] = dataset
    avg["target"] = target
    avg["model"] = model
    return avg[["dataset", "target", "model", "layer", "RSA", "CKA"]]


def load_tvsd_subjects(path, roi, split="test"):
    """Return list of (ids, Y) per monkey for the given ROI."""
    subjects_data = []
    with h5py.File(path, "r") as f:
        ids = f[f"{split}/stimulus_ids"][:]
        subjects = list(f[f"{split}/neural_data"].keys())
        for sub in subjects:
            if roi in f[f"{split}/neural_data/{sub}"]:
                data = f[f"{split}/neural_data/{sub}/{roi}"][:]
                subjects_data.append((ids, data))
    return subjects_data

def load_eeg_subjects(path, roi, tmin, tmax, split="test"):
    """Return list of (ids, Y) per subject for the given ROI and time window."""
    subjects_data = []
    with h5py.File(path, "r") as f:
        ids = f[f"{split}/stimulus_ids"][:]
        subjects = list(f[f"{split}/neural_data"].keys())
        for sub in subjects:
            data = f[f"{split}/neural_data/{sub}/{roi}"][:]
            data = data[:, :, tmin:tmax]
            data = data.reshape(data.shape[0], -1)
            subjects_data.append((ids, data))
    return subjects_data

def load_nsd_subjects(path, roi, split="test"):
    """Return list of (ids, Y) per subject for the given ROI."""
    subjects_data = []
    with h5py.File(path, "r") as f:
        subjects = list(f[f"{split}/stimulus_ids"].keys())
        for sub in subjects:
            ids = f[f"{split}/stimulus_ids/{sub}"][:]
            if roi in f[f"{split}/neural_data/{sub}"]:
                Y = f[f"{split}/neural_data/{sub}/{roi}"][:]
                subjects_data.append((ids, Y))
    return subjects_data
````

## 2.4 Predictive alignment: linear encoding models

## Linear encoding model

In the predictive part of the project, you will map model features to neural responses using a **linear encoding model with L2 regularization (ridge regression)**.

For a stimulus $\mathbf{x}$, let $\mathbf{z}_{\ell}(\mathbf{x})$ denote the feature vector extracted from model layer $\ell$. For a given subject $s$ and neural target $r$ (for example an ROI, a group of voxels, or a set of channels / time points), the predicted neural response is

$$
\widehat{\mathbf{y}}_{r,s}(\mathbf{x})
=
W_{r,s}\,\mathbf{z}_{\ell}(\mathbf{x}) + \mathbf{b}_{r,s},
$$

where:
- $\mathbf{z}_{\ell}(\mathbf{x}) \in \mathbb{R}^{d}$ is the model feature vector,
- $\widehat{\mathbf{y}}_{r,s}(\mathbf{x}) \in \mathbb{R}^{p}$ is the predicted neural response,
- $W_{r,s} \in \mathbb{R}^{p \times d}$ is the learned linear mapping,
- $\mathbf{b}_{r,s} \in \mathbb{R}^{p}$ is a bias term.

We fit the mapping on the training split using ridge regression:

$$
\min_{W_{r,s},\,\mathbf{b}_{r,s}}
\sum_{\mathbf{x}\in\mathcal{D}_{\mathrm{train}}}
\left\|
\mathbf{y}_{r,s}(\mathbf{x}) - \widehat{\mathbf{y}}_{r,s}(\mathbf{x})
\right\|_2^2
\;+\;
\alpha \left\|W_{r,s}\right\|_F^2.
$$

Here, $\mathbf{y}_{r,s}(\mathbf{x})$ is the measured neural response, and $\alpha$ controls the strength of L2 regularization. Larger $\alpha$ penalizes large weights more strongly, which can improve generalization when the feature dimension is high. You should select $\alpha$ using only the training data, for example with a validation split or cross-validation, and then evaluate the final model on the held-out test set.

**Define train/val/test procedure**

````python
from sklearn.linear_model import RidgeCV
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import warnings
from concurrent.futures import ThreadPoolExecutor

MAX_PCA   = 512
ALPHAS    = np.logspace(-2, 6, 17)   # 17 candidates: 0.01 … 1e6
NC_THRESH = 0.1                      # EEG reliability filter threshold


def pearsonr_vec(yt, yp):
    """Vectorised per-unit Pearson r → shape (n_units,)."""
    yt = yt - yt.mean(0);  yp = yp - yp.mean(0)
    num   = (yt * yp).sum(0)
    denom = np.sqrt((yt**2).sum(0) * (yp**2).sum(0)) + 1e-12
    return np.clip(num / denom, -1.0, 1.0)

def ev_vec(yt, yp):
    """Vectorised per-unit explained variance → shape (n_units,)."""
    ss_res = ((yt - yp)**2).sum(0)
    ss_tot = ((yt - yt.mean(0))**2).sum(0) + 1e-12
    return np.clip(1.0 - ss_res / ss_tot, -1.0, 1.0)

def nc_clip(scores, nc):
    """Noise-ceiling correction, clipped to [0, 1]."""
    return np.clip(scores / np.maximum(nc, 1e-8), 0.0, 1.0)

def agg(arr, nc=None, eeg=False):
    """Mean over units, returning a single scalar per layer per model.

    Aggregating to a scalar is necessary to rank layers, compare models,
    and populate the summary DataFrame fed into best_layer_stats.

    For EEG, a noise-ceiling mask (nc > NC_THRESH) is applied before averaging
    to exclude unreliable channel-time units — e.g. artifact-contaminated channels
    or time points with no evoked response. Without this mask, noisy units would
    drag the mean toward zero even when the model predicts reliable units well.
    NC_THRESH=0.1 retains only units where at least 10% of the variance is real signal.
    """
    if eeg and nc is not None:
        mask = nc > NC_THRESH
        return float(arr[mask].mean()) if mask.any() else np.nan
    return float(np.nanmean(arr))


_rsa = RepresentationalSimilarityAnalysis()
_cka = CenteredKernelAlignment()
METRIC_COLS = ['pearsonr', 'pearsonr_nc', 'ev', 'ev_nc', 'enc_rsa', 'enc_cka']


def precompute_alignment(feature_path, neural_ids):
    """Compute sorted row indices and restore order without loading features.

    Should be called once per dataset split rather than once per layer, since
    the ID-to-index mapping is identical across all layers of the same file.
    Returns (sorted_idx, restore_order) to enable fast contiguous I/O in
    load_layer_fast.
    """
    with h5py.File(feature_path, "r") as f:
        feat_ids = f["ids"][:].astype(str)
    neural_ids_str = np.array(neural_ids).astype(str)
    feat_id_to_idx = {v: i for i, v in enumerate(feat_ids)}
    missing = [x for x in neural_ids_str if x not in feat_id_to_idx]
    if missing:
        raise KeyError(f"{len(missing)} IDs not found: {missing[:5]}")
    feat_idx = np.array([feat_id_to_idx[x] for x in neural_ids_str])
    sort_order = np.argsort(feat_idx)
    restore = np.empty_like(sort_order)
    restore[sort_order] = np.arange(len(sort_order))
    return feat_idx[sort_order], restore


def load_layer_fast(feature_path, layer, sorted_idx, restore):
    """Load all rows with a single sequential HDF5 read, then reorder in RAM.

    Avoids per-row random seeks by reading contiguous sorted indices in one I/O
    call, then restoring the original neural ordering via the restore array
    precomputed in precompute_alignment.
    """
    with h5py.File(feature_path, "r") as f:
        X = f[layer][sorted_idx]
    return X[restore].astype(np.float32)


def preprocess_once(X_tr, X_te):
    """Scale and optionally PCA-reduce features, fit on train only.

    Called once per (model, layer) rather than per target to avoid redundant
    computation. If the feature dimension exceeds MAX_PCA, PCA is applied to
    keep the ridge regression tractable.
    """
    sc = StandardScaler()
    Xtr = sc.fit_transform(X_tr)
    Xte = sc.transform(X_te)
    if Xtr.shape[1] > MAX_PCA:
        pca = PCA(n_components=min(MAX_PCA, Xtr.shape[0] - 1), random_state=0)
        Xtr = pca.fit_transform(Xtr)
        Xte = pca.transform(Xte)
    return Xtr, Xte


def fit_and_eval(Xtr, y_tr, Xte, y_te, nc, is_eeg=False):
    """Fit a Ridge regression and evaluate predictions on the test set.

    Expects already-scaled and PCA-reduced features from preprocess_once.
    Uses RidgeCV with Generalised Cross-Validation (cv=None) instead of k-fold,
    which is a closed-form efficient solution — critical for EEG with its large
    output dimension. Returns per-unit Pearson r, explained variance, their
    noise-ceiling-corrected versions, and RSA/CKA between actual and predicted
    responses.
    """
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        ridge = RidgeCV(alphas=ALPHAS, cv=None, fit_intercept=True)
        ridge.fit(Xtr, y_tr)
    y_pred = ridge.predict(Xte)
    r      = pearsonr_vec(y_te, y_pred)
    ev     = ev_vec(y_te, y_pred)
    nc_corr = np.sqrt(np.maximum(nc, 0.0))
    kw = dict(nc=nc, eeg=is_eeg)
    return {
        "pearsonr":    agg(r, **kw),
        "pearsonr_nc": agg(nc_clip(r, nc_corr), **kw),
        "ev":          agg(ev, **kw),
        "ev_nc":       agg(nc_clip(ev, nc), **kw),
        "enc_rsa":     _rsa(y_te, y_pred),
        "enc_cka":     _cka(y_te, y_pred),
        "best_alpha":  ridge.alpha_,
    }
````

**Fit linear models across layers and compute predictive metrics**

````python
# ── TVSD ──────────────────────────────────────────────────────────────────────
rows_tvsd = []
print("=== TVSD ===")
with h5py.File(TVSD_path, "r") as f:
    tvsd_train_ids = f["train/stimulus_ids"][:]
    tvsd_test_ids  = f["test/stimulus_ids"][:]

for model_name, paths in models.items():
    layers = get_feature_layers(paths["THINGS"])
    print(f"  [{model_name}] {len(layers)} layers")
    tr_idx, tr_restore = precompute_alignment(paths["THINGS"], tvsd_train_ids)
    te_idx, te_restore = precompute_alignment(paths["THINGS"], tvsd_test_ids)

    for lyr in tqdm(layers, desc=model_name):
        X_tr_raw = load_layer_fast(paths["THINGS"], lyr, tr_idx, tr_restore)
        X_te_raw = load_layer_fast(paths["THINGS"], lyr, te_idx, te_restore)
        Xtr, Xte = preprocess_once(X_tr_raw, X_te_raw)
        del X_tr_raw, X_te_raw

        for tgt in tvsd_train:
            # tgt is 'monkeyF/V1' etc. — split into subject and roi
            monkey, roi = tgt.split('/')
            y_tr = tvsd_train[tgt]
            y_te = tvsd_test[tgt]
            nc   = tvsd_nc[tgt]
            res  = fit_and_eval(Xtr, y_tr, Xte, y_te, nc)
            rows_tvsd.append({
                "dataset": "TVSD",
                "model":   model_name,
                "layer":   lyr,
                "target":  roi,      # V1, V4, IT
                "subject": monkey,   # monkeyF, monkeyN
                **{k: res[k] for k in METRIC_COLS}
            })

df_TVSD = pd.DataFrame(rows_tvsd)
df_TVSD.to_csv("df_TVSD.csv", index=False)
print(f"Saved df_TVSD.csv: {len(df_TVSD)} rows")
print(df_TVSD[['model', 'target', 'subject']].drop_duplicates().sort_values(['model', 'target', 'subject']))

# ── EEG2 ──────────────────────────────────────────────────────────────────────
print("\n=== EEG2 ===")

EEG2_ROI = "occipital_parietal"

time_windows = {
    "early_0_200ms":    (0,  20),
    "middle_200_500ms": (20, 50),
    "late_500_800ms":   (50, 80),
}

with h5py.File(EEG2_path, "r") as f:
    eeg2_train_ids = f["train/stimulus_ids"][:]
    eeg2_test_ids  = f["test/stimulus_ids"][:]
    eeg2_subs = sorted(f["train/neural_data"].keys())

    # Load raw 3D (n, ch, t) — do NOT flatten yet
    eeg2_train_raw, eeg2_test_raw, eeg2_nc_raw = {}, {}, {}
    for sub in eeg2_subs:
        eeg2_train_raw[sub] = f[f"train/neural_data/{sub}/{EEG2_ROI}"][:]
        eeg2_test_raw[sub]  = f[f"test/neural_data/{sub}/{EEG2_ROI}"][:]
        eeg2_nc_raw[sub]    = f[f"noise_ceilings/{sub}/{EEG2_ROI}"][:] / 100.0

rows_eeg2 = []

for model_name, paths in models.items():
    layers = get_feature_layers(paths["THINGS"])
    print(f"  [{model_name}] {len(layers)} layers")
    tr_idx, tr_restore = precompute_alignment(paths["THINGS"], eeg2_train_ids)
    te_idx, te_restore = precompute_alignment(paths["THINGS"], eeg2_test_ids)

    for lyr in tqdm(layers, desc=model_name):
        X_tr_raw = load_layer_fast(paths["THINGS"], lyr, tr_idx, tr_restore)
        X_te_raw = load_layer_fast(paths["THINGS"], lyr, te_idx, te_restore)
        Xtr, Xte = preprocess_once(X_tr_raw, X_te_raw)
        del X_tr_raw, X_te_raw

        for win_name, (tmin, tmax) in time_windows.items():
            for sub in eeg2_subs:
                raw_tr = eeg2_train_raw[sub][:, :, tmin:tmax]
                raw_te = eeg2_test_raw[sub][:, :, tmin:tmax]
                nc     = eeg2_nc_raw[sub][:, tmin:tmax].ravel()

                y_tr = raw_tr.reshape(raw_tr.shape[0], -1)  # (n, ch*t)
                y_te = raw_te.reshape(raw_te.shape[0], -1)

                res = fit_and_eval(Xtr, y_tr, Xte, y_te, nc, is_eeg=True)
                rows_eeg2.append({
                    "dataset": "EEG2",
                    "model":   model_name,
                    "layer":   lyr,
                    "target":  win_name,
                    "subject": sub,
                    **{k: res[k] for k in METRIC_COLS}
                })

df_EEG2 = pd.DataFrame(rows_eeg2)
df_EEG2.to_csv("df_EEG2.csv", index=False)
print(f"Saved df_EEG2.csv: {len(df_EEG2)} rows")
print(df_EEG2[["model", "target", "subject"]].drop_duplicates().sort_values(["model", "target", "subject"]))

# ── NSD ───────────────────────────────────────────────────────────────────────
rows=[]
print("\n=== NSD ===")
for model_name, paths in models.items():
    layers = get_feature_layers(paths["NSD"])
    print(f"  [{model_name}] {len(layers)} layers")

    for sub in tqdm(nsd_train, desc=model_name):
        with h5py.File(NSD_path, "r") as f:
            nsd_train_ids = f[f"train/stimulus_ids/{sub}"][:]
            nsd_test_ids  = f[f"test/stimulus_ids/{sub}"][:]

        # Alignment computed once per subject (IDs differ per subject in NSD)
        tr_idx, tr_restore = precompute_alignment(paths["NSD"], nsd_train_ids)
        te_idx, te_restore = precompute_alignment(paths["NSD"], nsd_test_ids)

        for lyr in layers:
            X_tr_raw = load_layer_fast(paths["NSD"], lyr, tr_idx, tr_restore)
            X_te_raw = load_layer_fast(paths["NSD"], lyr, te_idx, te_restore)

            # PCA once per (subject, layer) — shared across all NSD ROIs
            Xtr, Xte = preprocess_once(X_tr_raw, X_te_raw)
            del X_tr_raw, X_te_raw

            for roi in nsd_train[sub]:
                y_tr = nsd_train[sub][roi]; y_te = nsd_test[sub][roi]
                nc   = nsd_nc[sub][roi]
                res  = fit_and_eval(Xtr, y_tr, Xte, y_te, nc)
                rows.append({"dataset": "NSD", "model": model_name,
                             "layer": lyr, "target": roi, "subject": sub,
                             **{k: res[k] for k in METRIC_COLS}})

df_NSD = pd.DataFrame(rows)
df_NSD.to_csv("df_NSD.csv", index=False)
print(f"Saved df_NSD.csv: {len(df_NSD)} rows")
print(df_NSD[["model", "target", "subject"]].drop_duplicates().sort_values(["model", "target", "subject"]))
````

**Layer wise plots (averaged over targets/subjects)**

````python
import re

datasets = {
    'TVSD': pd.read_csv("df_TVSD.csv"),
    'EEG2': pd.read_csv("df_EEG2.csv"),
    'NSD':  pd.read_csv("df_NSD.csv"),
}

def sort_layers(layers):
    def key(name):
        n = name.split('/')[-1] if '/' in name else name
        m = re.match(r'layer(\d+)-(\d+)$', n)
        if m:
            return (0, int(m.group(1)), int(m.group(2)))
        m = re.match(r'visual-blocks-(\d+)$', n)
        if m:
            return (1, int(m.group(1)), 0)
        m = re.match(r'language_model-layers-(\d+)$', n)
        if m:
            return (2, int(m.group(1)), 0)
        return (99, 0, 0)
    return sorted(layers, key=key)

ALL_METRICS = ['pearsonr', 'pearsonr_nc', 'ev', 'ev_nc', 'enc_rsa', 'enc_cka']

for ds_name, df in datasets.items():
    for mdl in df['model'].unique():
        sub          = df[df['model'] == mdl].copy()
        layer_order  = sort_layers(sub['layer'].unique())
        lyr_idx      = {l: i for i, l in enumerate(layer_order)}
        layer_labels = [l.split('/')[-1] for l in layer_order]

        fig, ax = plt.subplots(figsize=(10, 4))

        for metric in ALL_METRICS:
            grp = sub.groupby('layer')[metric].mean().reset_index()
            grp['li'] = grp['layer'].map(lyr_idx)
            grp = grp.sort_values('li')
            ax.plot(grp['li'], grp[metric], marker='o', lw=1.4, ms=4, label=metric)

        ax.set_xticks(range(len(layer_labels)))
        ax.set_xticklabels(layer_labels, rotation=60, ha='right', fontsize=7)
        ax.set_xlim(-0.5, len(layer_labels) - 0.5)
        ax.legend(fontsize=8, bbox_to_anchor=(1.01, 1), loc='upper left')
        ax.grid(axis='y', alpha=0.3, lw=0.5)
        ax.set_title(f'Layer-wise {ds_name} - {mdl} - avg over targets', fontsize=10)
        plt.tight_layout()
        plt.savefig(f'layerwise_{ds_name}_{mdl}.png', dpi=150, bbox_inches='tight')
        plt.show()
````

````python
# DELIVERABLE 3 — Time-resolved EEG layer comparison
# For each time window, plot the full layer-wise CKA profile.
# A rightward shift of the peak across windows (early → middle → late)
# indicates that later processing stages recruit deeper model representations.
r26 = pd.read_csv("results_23.csv")
eeg = r26[r26["dataset"] == "EEG2"]

# pearsonr_nc EEG data: average across subjects per (model, target, layer)
eeg_enc = (df_enc[df_enc["dataset"] == "EEG2"]
           .groupby(["model", "target", "layer"])["pearsonr_nc"]
           .mean()
           .reset_index())

eeg_cka = r26[r26["dataset"] == "EEG2"]

metric_sources_d3 = [
    ("CKA",         eeg_cka, "CKA"),
    ("pearsonr_nc", eeg_enc, "pearsonr_nc"),
]

model_names = sorted(eeg_cka["model"].unique())
n_models    = len(model_names)

fig, axes = plt.subplots(2, n_models, figsize=(9 * n_models, 10), sharey=False)
axes = np.array(axes).reshape(2, n_models)

for row_i, (metric_label, df_src, metric_col) in enumerate(metric_sources_d3):
    for col_i, model_name in enumerate(model_names):
        ax  = axes[row_i, col_i]
        sub = df_src[df_src["model"] == model_name]

        all_layers = sort_layers(sub["layer"].unique().tolist())
        short_lbl  = [l.split("/")[-1] for l in all_layers]
        x          = range(len(all_layers))

        for window in EEG_TIME_ORDER:
            win_data = sub[sub["target"] == window]
            if win_data.empty:
                continue
            vals = [win_data.loc[win_data["layer"] == l, metric_col].mean()
                    if l in win_data["layer"].values else float("nan")
                    for l in all_layers]
            col   = WINDOW_COLORS[window]
            ax.plot(x, vals, "o-", color=col, lw=1.8, ms=5, label=WINDOW_LABELS[window])
            peak_i = int(np.nanargmax(vals))
            ax.axvline(peak_i, color=col, lw=0.8, linestyle="--", alpha=0.6)

        ax.set_xticks(list(x))
        ax.set_xticklabels(short_lbl, rotation=45, ha="right", fontsize=7)
        ax.set_xlabel("Model layer  (shallow -> deep)", fontsize=10)
        ax.set_ylabel(metric_label, fontsize=10)
        ax.set_title(f"{model_name}  —  {metric_label}", fontsize=11)
        ax.legend(fontsize=9)
        ax.grid(axis="y", alpha=0.3)

plt.suptitle("Time-resolved EEG layer comparison  (EEG2)", fontsize=13, fontweight="bold")
plt.tight_layout()
plt.savefig("images/2_6_eeg_timeresolved_layers.png", dpi=150, bbox_inches="tight")
plt.show()

# Summary table: peak layer per window and metric
print("\n-- Peak layer per EEG time window --")
for metric_label, df_src, metric_col in metric_sources_d3:
    print(f"\n  [{metric_label}]")
    for model_name in model_names:
        sub        = df_src[df_src["model"] == model_name]
        all_layers = sort_layers(sub["layer"].unique().tolist())
        print(f"    {model_name}")
        for window in EEG_TIME_ORDER:
            win_data = sub[sub["target"] == window]
            if win_data.empty:
                continue
            best_l = win_data.loc[win_data[metric_col].idxmax(), "layer"]
            best_v = win_data[metric_col].max()
            depth  = all_layers.index(best_l) / max(len(all_layers) - 1, 1)
            print(f"      {window:25s}  best: {best_l.split('/')[-1]:30s}  "
                  f"depth={depth:.2f}  {metric_label}={best_v:.4f}")
````

**Hypothesis**

We test whether sharing information across subjects can improve EEG prediction. The hypothesis is that averaging neural responses across subjects reduces noise and enhances the shared stimulus-evoked component. However, we also hypothesize that pure pooling may discard subject-specific information, and that combining a shared component with subject-specific readouts can better capture both shared and individual variability. As a further extension, we hypothesize that explicitly aligning each subject's response subspace into a common latent space before pooling, using PCA and orthogonal Procrustes rotation, can produce a cleaner group average than normalization alone, and that adding subject-specific residuals on top of this aligned model can yield additional gains.

**Motivation** 

EEG signals are inherently noisy, and in this dataset each image is observed only once per subject. This makes the per-subject training regime particularly sensitive to noise. Pooling across subjects is a natural strategy to increase the effective signal-to-noise ratio, under the assumption that stimulus-driven responses are consistent across individuals. However, EEG signals also exhibit strong inter-subject variability in amplitude, timing, and spatial topography, which may limit the effectiveness of simple averaging. Amplitude differences can be addressed by normalization, but topographic differences, arising from individual anatomy and electrode placement, require a more principled alignment of the response spaces before combining subjects. This motivates aligning each subject's neural subspace to a shared reference frame prior to pooling, and then correcting for residual individual variability through subject-specific readouts.

**Methods** 

The baseline model consists of a ridge regression readout trained independently for each subject, using features from a single model layer to predict EEG responses. In this extension, we evaluate five strategies, forming a progressive hierarchy:

- Pooled (raw): EEG responses are averaged directly across subjects.
- Pooled (z-scored): each subject's EEG responses are normalized (zero mean, unit variance per unit) before averaging, to reduce amplitude differences across subjects.
- Shared + subject-specific residuals: a shared Ridge model is first trained on the z-scored group average to capture the common stimulus-driven component; then for each subject a second Ridge is trained on the residuals (individual responses minus shared prediction), adding a subject-specific correction.
- Latent-space aligned: each subject's training responses are projected into a K=17-dimensional PCA subspace where K = n_channels (the number of EEG electrodes in the ROI), capturing the full spatial rank of the data while discarding redundant temporal components, then orthogonal Procrustes rotation is applied to align each subject's subspace to a shared reference computed from the group average. Responses are averaged in this common space before fitting a single shared Ridge model.
- Latent-aligned + subject-specific residuals: the same residual correction as method 3 is applied on top of the Procrustes-aligned shared model, combining both alignment and individual readout in a single pipeline.

All models are evaluated on each individual subject's test data using noise-corrected Pearson correlation, across three temporal windows (early: 0–200 ms, middle: 200–500 ms, late: 500–800 ms).

````python
# ============================================================
# Point 3 — Cross-subject pooling for EEG2
# ============================================================
# Six methods compared against the per-subject baseline:
#
#  1. per_subject_baseline     : Ridge( X, Y_s )
#  2. pooled_raw               : Ridge( X, mean(Y_s) )
#  3. pooled_zscored           : Ridge( X, mean(zscore(Y_s)) )
#  4. shared_plus_residual     : stage-1 Ridge on z-scored mean +
#                                per-subject Ridge on residuals
#  5. latent_aligned           : Procrustes-align each subject's PCA
#                                subspace to shared reference, average
#                                in common space, then Ridge
#  6. latent_aligned_residual  : same as 5 + per-subject residual Ridge
#
# All methods use the same DNN features (best layer from 2.4),
# the same Ridge pipeline, and the same per-subject NC evaluation.
# ============================================================

import numpy as np, pandas as pd, matplotlib.pyplot as plt
import h5py, warnings
from sklearn.linear_model import RidgeCV
from sklearn.decomposition import PCA
from scipy.linalg import svd as la_svd

EEG2_ROI    = "occipital_parietal"
ALPHAS      = np.logspace(-2, 6, 17)
METRIC_COLS = ["pearsonr", "pearsonr_nc", "ev", "ev_nc"]

time_windows = {
    "early_0_200ms":    (0,  20),
    "middle_200_500ms": (20, 50),
    "late_500_800ms":   (50, 80),
}

MODELS = {"ResNet": RN_EEG_path, "Transformer": T_EEG_path}

# ── Load EEG data ─────────────────────────────────────────────
with h5py.File(EEG2_path, "r") as f:
    eeg2_subs      = sorted(f["train/neural_data"].keys())
    eeg2_train_ids = f["train/stimulus_ids"][:]
    eeg2_test_ids  = f["test/stimulus_ids"][:]
    eeg2_train_raw, eeg2_test_raw, eeg2_nc_raw = {}, {}, {}
    for sub in eeg2_subs:
        eeg2_train_raw[sub] = f[f"train/neural_data/{sub}/{EEG2_ROI}"][:]
        eeg2_test_raw[sub]  = f[f"test/neural_data/{sub}/{EEG2_ROI}"][:]
        eeg2_nc_raw[sub]    = f[f"noise_ceilings/{sub}/{EEG2_ROI}"][:] / 100.0

n_train = eeg2_train_raw[eeg2_subs[0]].shape[0]
n_ch    = eeg2_train_raw[eeg2_subs[0]].shape[1]

print(f"Loaded {len(eeg2_subs)} subjects | ROI: {EEG2_ROI}")
print(f"Train shape per subject: {eeg2_train_raw[eeg2_subs[0]].shape}")
print(f"Test  shape per subject: {eeg2_test_raw[eeg2_subs[0]].shape}")

if "df_EEG2" not in vars():
    df_EEG2 = pd.read_csv("df_EEG2.csv")

# ── Helpers ───────────────────────────────────────────────────
def eval_row(model_name, win_name, sub, method, y_te, y_pred, nc):
    nc_corr = np.sqrt(np.maximum(nc, 0.0))
    r  = pearsonr_vec(y_te, y_pred)
    ev = ev_vec(y_te, y_pred)
    return {
        "model":       model_name,
        "target":      win_name,
        "subject":     sub,
        "method":      method,
        "pearsonr":    agg(r),
        "pearsonr_nc": agg(nc_clip(r, nc_corr), nc=nc, eeg=True),
        "ev":          agg(ev),
        "ev_nc":       agg(nc_clip(ev, nc),     nc=nc, eeg=True),
    }

def procrustes_align(V_s, V_ref):
    """Orthogonal Procrustes: find rotation R minimising ||V_ref - V_s @ R||_F."""
    U, _, Vt = la_svd(V_s.T @ V_ref, full_matrices=False)
    return U @ Vt

def fit_ridge(Xtr, Y):
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        m = RidgeCV(alphas=ALPHAS, cv=5, fit_intercept=True)
        m.fit(Xtr, Y)
    return m
# ── PCA dimensionality selection: scree plot for K = n_ch ─────────────────────
# K = n_ch (number of EEG channels) is chosen as the natural spatial rank:
# the data has n_ch spatial sensors whose signals can combine into at most
# n_ch independent spatial modes. The scree plot below confirms that K = n_ch
# sits at or beyond the natural elbow of the variance curve, and that adding
# more components yields negligible additional signal.

_tmin, _tmax = 0, 20   # early window for the diagnostic
_Y_per_sub_z_diag = {}
for _sub in eeg2_subs:
    _y = eeg2_train_raw[_sub][:, :, _tmin:_tmax].reshape(n_train, -1)
    _mu, _sig = _y.mean(0), _y.std(0) + 1e-8
    _Y_per_sub_z_diag[_sub] = (_y - _mu) / _sig

_Y_mean_z_diag = np.mean(list(_Y_per_sub_z_diag.values()), axis=0)

# fit PCA up to 4×n_ch so the elbow is visible in context
_K_max  = min(n_ch * 4, _Y_mean_z_diag.shape[1], n_train - 1)
_pca_d  = PCA(n_components=_K_max, random_state=0).fit(_Y_mean_z_diag)
_cumvar = np.cumsum(_pca_d.explained_variance_ratio_) * 100

_K_chosen = n_ch

fig, ax = plt.subplots(figsize=(7, 3.5))
ax.plot(range(1, _K_max + 1), _cumvar, color='steelblue', lw=2)
ax.axvline(_K_chosen, color='tomato', ls='--', lw=1.8,
           label=f'K = n_ch = {_K_chosen}')
ax.axhline(_cumvar[_K_chosen - 1], color='tomato', ls=':', lw=1.2)
ax.scatter([_K_chosen], [_cumvar[_K_chosen - 1]], color='tomato', zorder=5, s=60,
           label=f'{_cumvar[_K_chosen-1]:.1f}% variance explained at K = n_ch')
ax.set_xlabel('Number of PCA components', fontsize=11)
ax.set_ylabel('Cumulative explained variance (%)', fontsize=11)
ax.set_title('Scree plot — group-averaged EEG (early window, occipital_parietal)', fontsize=10)
ax.legend(fontsize=9)
ax.set_xlim(0, _K_max + 1)
ax.set_ylim(0, 102)
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig("images/k.png")
plt.show()

print(f"K = n_ch = {_K_chosen}  -->  variance explained: {_cumvar[_K_chosen-1]:.1f}%")
print(f"Components to reach 80%: {np.searchsorted(_cumvar, 80) + 1}")
print(f"Components to reach 90%: {np.searchsorted(_cumvar, 90) + 1}")
print(f"Components to reach 95%: {np.searchsorted(_cumvar, 95) + 1}")

del _Y_per_sub_z_diag, _Y_mean_z_diag, _pca_d, _cumvar

# ── Main loop ─────────────────────────────────────────────────
rows = []

for model_name, feature_path in MODELS.items():

    tr_idx, tr_restore = precompute_alignment(feature_path, eeg2_train_ids)
    te_idx, te_restore = precompute_alignment(feature_path, eeg2_test_ids)

    best_layer = (
        df_EEG2[df_EEG2["model"] == model_name]
        .groupby("layer")["pearsonr_nc"].mean().idxmax()
    )
    print(f"\n[{model_name}] best layer: {best_layer}")

    X_tr_raw = load_layer_fast(feature_path, best_layer, tr_idx, tr_restore)
    X_te_raw = load_layer_fast(feature_path, best_layer, te_idx, te_restore)
    Xtr, Xte = preprocess_once(X_tr_raw, X_te_raw)
    del X_tr_raw, X_te_raw

    for win_name, (tmin, tmax) in time_windows.items():

        # ── Per-subject training targets ──────────────────────
        Y_per_sub, Y_per_sub_z = {}, {}
        for sub in eeg2_subs:
            y = eeg2_train_raw[sub][:, :, tmin:tmax].reshape(n_train, -1)
            Y_per_sub[sub] = y
            mu = y.mean(0);  sigma = y.std(0) + 1e-8
            Y_per_sub_z[sub] = (y - mu) / sigma

        Y_tr_mean   = np.mean(list(Y_per_sub.values()),   axis=0)
        Y_tr_mean_z = np.mean(list(Y_per_sub_z.values()), axis=0)

        # ── Latent-space alignment (Procrustes) ───────────────
        # K = number of EEG channels: natural spatial rank of the data
        K = min(n_ch, Y_tr_mean_z.shape[1], n_train - 1)

        # Reference subspace from the z-scored group average
        pca_ref = PCA(n_components=K, random_state=0)
        pca_ref.fit(Y_tr_mean_z)
        V_ref = pca_ref.components_.T          # (n_units, K)

        # Align each subject's subspace to V_ref, project, reconstruct
        Y_aligned_list = []
        for sub in eeg2_subs:
            y_z = Y_per_sub_z[sub]
            pca_s = PCA(n_components=K, random_state=0)
            pca_s.fit(y_z)
            V_s = pca_s.components_.T          # (n_units, K)

            R_s       = procrustes_align(V_s, V_ref)
            coords    = y_z @ V_s @ R_s        # (n_train, K)
            y_aligned = coords @ V_ref.T       # (n_train, n_units)
            Y_aligned_list.append(y_aligned)

        Y_mean_aligned = np.mean(Y_aligned_list, axis=0)

        var_exp = pca_ref.explained_variance_ratio_.sum()
        print(f"  [{win_name}] K={K}, ref var explained={var_exp:.1%}")

        # ── Fit all pooled models ─────────────────────────────
        ridge_raw     = fit_ridge(Xtr, Y_tr_mean)
        ridge_z       = fit_ridge(Xtr, Y_tr_mean_z)
        ridge_aligned = fit_ridge(Xtr, Y_mean_aligned)

        Y_pred_raw            = ridge_raw.predict(Xte)
        Y_pred_z              = ridge_z.predict(Xte)
        Y_pred_aligned        = ridge_aligned.predict(Xte)

        # Shared training predictions (for residual stages)
        Y_pred_tr_z       = ridge_z.predict(Xtr)
        Y_pred_tr_aligned = ridge_aligned.predict(Xtr)

        # ── Per-subject evaluation ────────────────────────────
        for sub in eeg2_subs:
            y_tr = Y_per_sub[sub]
            y_te = eeg2_test_raw[sub][:, :, tmin:tmax].reshape(
                       eeg2_test_raw[sub].shape[0], -1)
            nc   = eeg2_nc_raw[sub][:, tmin:tmax].ravel()

            # 1. Per-subject baseline
            res_b = fit_and_eval(Xtr, y_tr, Xte, y_te, nc, is_eeg=True)
            rows.append({
                "model": model_name, "target": win_name,
                "subject": sub, "method": "per_subject_baseline",
                **{k: res_b[k] for k in METRIC_COLS}
            })

            # 2. Pooled raw
            rows.append(eval_row(model_name, win_name, sub,
                                 "pooled_raw", y_te, Y_pred_raw, nc))

            # 3. Pooled z-scored
            rows.append(eval_row(model_name, win_name, sub,
                                 "pooled_zscored", y_te, Y_pred_z, nc))

            # 4. Shared (z-scored) + per-subject residual
            ridge_res_z = fit_ridge(Xtr, y_tr - Y_pred_tr_z)
            Y_pred_shared_res = Y_pred_z + ridge_res_z.predict(Xte)
            rows.append(eval_row(model_name, win_name, sub,
                                 "shared_plus_residual", y_te,
                                 Y_pred_shared_res, nc))

            # 5. Latent-space aligned pooling
            rows.append(eval_row(model_name, win_name, sub,
                                 "latent_aligned", y_te, Y_pred_aligned, nc))

            # 6. Latent-aligned + per-subject residual
            ridge_res_al = fit_ridge(Xtr, y_tr - Y_pred_tr_aligned)
            Y_pred_aligned_res = Y_pred_aligned + ridge_res_al.predict(Xte)
            rows.append(eval_row(model_name, win_name, sub,
                                 "latent_aligned_residual", y_te,
                                 Y_pred_aligned_res, nc))

        print(f"  {win_name} ✓")

df_all = pd.DataFrame(rows)
df_all.to_csv("df_eeg_all_methods.csv", index=False)
print(f"\n✓ Saved df_eeg_all_methods.csv ({len(df_all)} rows)")

# ── Summary table ─────────────────────────────────────────────
METHOD_ORDER = [
    "per_subject_baseline",
    "pooled_raw",
    "pooled_zscored",
    "shared_plus_residual",
    "latent_aligned",
    "latent_aligned_residual",
]

pivot = (
    df_all.groupby(["model", "target", "method"])["pearsonr_nc"]
    .mean().unstack("method")[METHOD_ORDER].reset_index()
)
for col, dcol in [
    ("pooled_raw",             "Δ_raw"),
    ("pooled_zscored",         "Δ_zscored"),
    ("shared_plus_residual",   "Δ_shared"),
    ("latent_aligned",         "Δ_aligned"),
    ("latent_aligned_residual","Δ_aligned_res"),
]:
    pivot[dcol] = (pivot[col] - pivot["per_subject_baseline"]).round(4)

print("\n── pearsonr_nc — all six methods (avg over 10 subjects) ──")
display(pivot[["model", "target",
               "per_subject_baseline",
               "pooled_raw", "pooled_zscored",
               "shared_plus_residual",
               "latent_aligned", "latent_aligned_residual",
               "Δ_raw", "Δ_zscored", "Δ_shared",
               "Δ_aligned", "Δ_aligned_res"]].round(4))

# ── Plot ──────────────────────────────────────────────────────
wins       = list(time_windows.keys())
win_labels = ["Early\n0–200ms", "Middle\n200–500ms", "Late\n500–800ms"]

colors = {
    "per_subject_baseline":    "steelblue",
    "pooled_raw":              "tomato",
    "pooled_zscored":          "darkorange",
    "shared_plus_residual":    "forestgreen",
    "latent_aligned":          "mediumpurple",
    "latent_aligned_residual": "darkviolet",
}
labels = {
    "per_subject_baseline":    "Per-subject baseline",
    "pooled_raw":              "Pooled (raw)",
    "pooled_zscored":          "Pooled (z-scored)",
    "shared_plus_residual":    "Shared + residual",
    "latent_aligned":          "Latent-aligned",
    "latent_aligned_residual": "Latent-aligned + residual",
}

fig, axes = plt.subplots(1, 2, figsize=(17, 5), sharey=True)

for ax, mdl in zip(axes, ["ResNet", "Transformer"]):
    sub_piv = pivot[pivot["model"] == mdl]
    n_m     = len(METHOD_ORDER)
    x       = np.arange(len(wins))
    w       = 0.13
    offsets = np.linspace(-(n_m - 1) / 2, (n_m - 1) / 2, n_m) * w

    for method, offset in zip(METHOD_ORDER, offsets):
        vals = [sub_piv.loc[sub_piv["target"] == t, method].values[0]
                for t in wins]
        ax.bar(x + offset, vals, w,
               label=labels[method], color=colors[method], alpha=0.85)

    # Delta annotations on all non-baseline bars
    delta_map = [
        ("pooled_raw",             "Δ_raw",        offsets[1]),
        ("pooled_zscored",         "Δ_zscored",    offsets[2]),
        ("shared_plus_residual",   "Δ_shared",     offsets[3]),
        ("latent_aligned",         "Δ_aligned",    offsets[4]),
        ("latent_aligned_residual","Δ_aligned_res",offsets[5]),
    ]
    for xi, t in enumerate(wins):
        for method, dcol, offset in delta_map:
            d   = sub_piv.loc[sub_piv["target"] == t, dcol].values[0]
            val = sub_piv.loc[sub_piv["target"] == t, method].values[0]
            c   = "darkgreen" if d >= 0 else "crimson"
            ax.text(xi + offset, val + 0.009,
                    f"{'+'if d>=0 else ''}{d:.3f}",
                    ha="center", va="bottom",
                    fontsize=6, color=c, fontweight="bold")

    ax.set_xticks(x)
    ax.set_xticklabels(win_labels, fontsize=9)
    ax.set_ylim(0, 1.05)
    ax.set_ylabel("pearsonr_nc", fontsize=11)
    ax.set_title(mdl, fontsize=13)
    ax.legend(fontsize=7, loc="upper right")

fig.suptitle(
    f"Six pooling strategies vs per-subject baseline — EEG2 ({EEG2_ROI})",
    fontsize=12)
plt.tight_layout()
plt.savefig("crosssubject_EEG2_full.png", dpi=150, bbox_inches="tight")
plt.show()
````

**Discussion**

Raw pooling performs worst in all conditions (Δ ≈ −0.17 early, −0.04 late), confirming that naive averaging is not effective. The main issue is amplitude heterogeneity: subjects with larger responses dominate the average.
Z-scored pooling improves over raw pooling in all cases, and even exceeds the baseline in the Transformer late window (+0.021). However, the gap remains large in early and middle windows (Δ ≈ −0.12 to −0.16), showing that normalization alone is not enough.
Latent alignment (PCA + Procrustes) does not improve performance. In some cases, it is slightly worse than z-scored pooling (e.g., ResNet early: 0.507 vs 0.522). This is likely due to noise: PCA components are estimated from single-trial data and are not stable, so the alignment step introduces additional error.
The best results are obtained with methods that include a residual per subject. Both shared + residual and aligned + residual outperform the baseline in all conditions (Δ ≈ +0.001 to +0.009). The two methods give almost identical results, showing that the residual step is the key component.
The probability of observing six positive improvements by chance is low (≈ 1.6%), supporting the robustness of the result.
Across time windows, the largest improvements are in the middle window, while gains are smaller in the late window. Early responses are more subject-specific, so the residual helps more. Late responses are more consistent across subjects, so pooling already works well.

Overall, these results show that cross-subject information is beneficial only when individual variability is explicitly modeled through subject-specific residuals, rather than averaged away.
